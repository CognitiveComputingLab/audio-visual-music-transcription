```python
os.chdir("/content/drive/MyDrive/Dissertation Code/Skipping-The-Frame-Level")

input_directory = "../OMAPS2/complete/mp4"
output_directory = "../OMAPS2/evaluation/audio-visual-fusion-tuned"

os.makedirs(output_directory, exist_ok=True)

# Iterate over each wav
for filename in os.listdir(input_directory):
    if filename.endswith(".mp4"):

        input_video_path = os.path.join(input_directory, filename)
        output_midi_path = os.path.join(output_directory, os.path.splitext(filename)[0] + ".mid")

        # Transkun command
        command = f"python -m transkun.transcribe {input_video_path} {output_midi_path} --segmentHopSize 24 --segmentSize 32 --device cuda"

        print(f"Processing {input_video_path}...")

        try:
            subprocess.run(command, check=True, shell=True)
            print(f"Successfully processed {input_video_path} to {output_midi_path} --device cuda")
        except subprocess.CalledProcessError as e:
            print(f"An error occurred while processing {input_video_path}: {e}")

print("DONE")
```
