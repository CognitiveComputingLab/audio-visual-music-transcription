```python
import os
import librosa
import torchlibrosa
from piano_transcription_inference import PianoTranscription, sample_rate, load_audio

os.chdir("/content/drive/MyDrive/Dissertation Code")

input_directory = "OMAPS2/complete/wav_converted"
output_directory = "OMAPS2/evaluation/bytedance"

os.makedirs(output_directory, exist_ok=True)

transcriptor = PianoTranscription(device='cuda')

for filename in os.listdir(input_directory):
    if filename.endswith(".wav"):
        input_audio_path = os.path.join(input_directory, filename)
        output_midi_path = os.path.join(output_directory, os.path.splitext(filename)[0] + ".mid")

        print(f"Processing {input_audio_path}...")

        audio, _ = librosa.load(input_audio_path, sr=sample_rate, mono=True)

        try:
            transcribed_dict = transcriptor.transcribe(audio, output_midi_path)
            print(f"Successfully processed {input_audio_path} to {output_midi_path}")
        except Exception as e:
            print(f"An error occurred while processing {input_audio_path}: {e}")

print("DONE")
```
